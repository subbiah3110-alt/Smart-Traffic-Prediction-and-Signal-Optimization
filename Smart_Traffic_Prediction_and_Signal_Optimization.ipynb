```python
import tensorflow as tf 
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.preprocessing import image
import numpy as np

DATA_DIR = r"C:\Users\Subbiah\Training\Capstone 2\Train_2"

train_datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2, 
    rotation_range=15, 
    width_shift_range=0.1, 
    height_shift_range=0.1, 
    horizontal_flip=True
)

test_datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2
)

train_generator = train_datagen.flow_from_directory(
    DATA_DIR,
    target_size=(128, 128),
    batch_size=32,
    class_mode='categorical',
    subset='training'
)

test_generator = test_datagen.flow_from_directory(
    DATA_DIR,
    target_size=(128, 128),
    batch_size=32,
    class_mode='categorical',
    subset='validation' 
)

```

```python
base_model = tf.keras.applications.MobileNetV2(   
    input_shape=(128, 128, 3),
    include_top=False, 
    weights='imagenet' 
    
)
base_model.trainable = False 

model = Sequential([
    base_model,
    GlobalAveragePooling2D(),
    Dense(64, activation='relu'),
    Dropout(0.3),
    Dense(3, activation='softmax') 
])

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

early_stop = EarlyStopping(
    monitor='val_loss', 
    patience=3,          
    restore_best_weights=True 
)

history = model.fit(
    train_generator,
    epochs=12,
    validation_data=test_generator,
    callbacks=[early_stop]
)

```

```python
model.save('traffic_density_model.h5')
np.save('class_indices.npy', train_generator.class_indices)

from tensorflow.keras.models import load_model
model = load_model('traffic_density_model.h5')
class_indices = np.load('class_indices.npy', allow_pickle=True).item()
idx_to_class = {v: k for k, v in class_indices.items()} 

def signal_controller(pred_class):
    pred_cleaned = str(pred_class).lower()
    if "heavy" in pred_cleaned:
        return {"green": 40, "red": 20}
    elif "medium" in pred_cleaned:
        return {"green": 30, "red": 30}
    elif "low" in pred_cleaned:
        return {"green": 20, "red": 40}


```

```python
img_path = r"C:\Users\Subbiah\OneDrive\Desktop\Traffic_Project\Images\heavy\images_799.jpg" 

img = image.load_img(img_path, target_size=(128, 128)) 
img_array = image.img_to_array(img) / 255.0 
img_array = np.expand_dims(img_array, axis=0) 

pred = model.predict(img_array) 
pred_class = idx_to_class[np.argmax(pred)] 
signal = signal_controller(pred_class)

print(f"\nPredicted density: {pred_class}")
print(f"Signal timings → Green: {signal['green']}s, Red: {signal['red']}s")
```
